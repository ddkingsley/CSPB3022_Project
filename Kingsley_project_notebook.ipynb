## Models

As stated above we will build a number of multiple linear regression models based on forward stepwise refinement. In forward stepwise refinement we build a model one feature at a time by finding the best feature. We start with a model using the single best feature, then build another model using both the next best feature given the inclusion of the first best, and so on. Below is this process. Please see explanations in the code comments.

```python
# First we construnct training and test data subsets.
train_real = real_est.loc[np.random.choice(real_est.index, size=150)]
test_real = real_est.loc[np.random.choice(real_est.index, size=150)]
```

```python
# Here the features to be used are added to a list. 
features = ['trans_date','age','dist_to_MRT','num_stores','lat','long']
```

```python
# The params function finds the next best parameter (feature) and returns its index in
# the features list and the parameters that make up the model (using k parameters). 
def params(cur_params=0):
    # find the first parameter
    if cur_params == 0:
        adj_r2 = []
        # iterates through the available features and picks the best
        for p in features:
            mod_train = smf.ols(formula=f'price~{p}', data=train_real).fit()
            adj_r2.append(mod_train.rsquared_adj)
        best_index = adj_r2.index(max(adj_r2))
        # pick best parameter
        param1 = features[best_index]
        return best_index, param1
    # else find the next best parameter
    else:
        adj_r2 = []
        # iterates through the available features and picks the best
        for p in features:
            mod_train = smf.ols(formula=f'price~{cur_params}+{p}', data=train_real).fit()
            adj_r2.append(mod_train.rsquared_adj)
        best_index = adj_r2.index(max(adj_r2))
        # build parameters to be used in the model
        new_params = f'{cur_params}+{features[best_index]}'
        return best_index, new_params
```

```python
# Forward stepwise refinement

train_models = []
train_adj_rsquared = []
test_models = []
test_adj_rsquared = []
cur_params = 0
num_features = len(features)

# finds the best model for each number of parameters (1-6)
for i in range(num_features):
    param_indx, cur_params = params(cur_params)
    features.pop(param_indx)
    # build models from training data
    train_model = smf.ols(formula=f'price~{cur_params}', data=train_real).fit()
    train_models.append(train_model)
    train_adj_rsquared.append(train_model.rsquared_adj)
    # build models from test data
    test_model = smf.ols(formula=f'price~{cur_params}', data=test_real).fit()
    test_models.append(test_model)
    test_adj_rsquared.append(test_model.rsquared_adj)
    
    print(f'Best {i+1} parameter model: price = {cur_params}')
```

## Results and Analysis

Above the model is iteratively trained and evaluated to improve performance through forward stepwise refinement. Below is a visualization of the resulting adjusted R_squared values of each successive model created above. The adjusted R_squared value increases if the added parameter improves the model signifacantly. 

```python
# Compare results from the multiple models.
# k represents the number of features used in the model.
k = [1,2,3,4,5,6]
plt.plot(k, train_adj_rsquared)
plt.plot(k, test_adj_rsquared, 'ro')
plt.xlabel('K')
plt.ylabel('Adjusted Rsquared');
```

From the set of models created it appears that adjusted R_squared levels off at 4 parameters or features (price = dist_to_MRT + num_stores + age + lat). Since there is no significant improvement by adding any more features, the final model below applies the best 4 feature model from above to the entire dataset. 

```python
# apply best model to full dataset and summarize the results and analysis
real_mod = smf.ols(formula='price~dist_to_MRT+num_stores+age+lat', data=real_est).fit()
real_mod.summary()
```

**Evaluation metrics**
- R_squared, the coefficient of multiple determination, is 0.617 in the final model indicating the data are close to the fitted regression line. This suggests it is an accurate model.
- The p-values of all of the parameters are less than 0.05. This suggests that all of the parameters used in the model are significant.
